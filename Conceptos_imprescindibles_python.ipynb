## üß± **1. Sintaxis y Sem√°ntica**

- **Sintaxis**: Conjunto de reglas que determinan c√≥mo debe escribirse correctamente el c√≥digo en Python.
- **Sem√°ntica**: Significado y comportamiento que tiene el c√≥digo al ejecutarse.

### ‚úÖ Ejemplo 1: Indentaci√≥n correcta

La **indentaci√≥n correcta** en Python es usar espacios o tabulaciones uniformes para organizar el c√≥digo, indicando bloques y jerarqu√≠as.

```python

if True:
    print("Bien")  # ‚úÖ correcto

# ‚ùå Incorrecto

print("Mal")  # Error de indentaci√≥n

```

üìå Python **usa espacios como parte de la sintaxis**, no soporta llaves `{}` como otros lenguajes.

## üßÆ **2. Tipos de Datos Nativos**

Los **tipos de datos nativos** son las categor√≠as b√°sicas de valores que Python reconoce y maneja directamente, como n√∫meros, texto y booleanos.

En Python, los principales **tipos de datos nativos** son:

- **int**: N√∫meros enteros (ej. `5`)
- **float**: N√∫meros decimales (ej. `3.14`)
- **bool**: Valores booleanos (`True` o `False`)
- **str**: Cadenas de texto (ej. `"Hola"`)
- **list**: Listas (ej. `[1, 2, 3]`)
- **tuple**: Tuplas (ej. `(1, 2, 3)`)
- **dict**: Diccionarios (ej. `{"nombre": "Ana"}`)
- **set**: Conjuntos (ej. `{1, 2, 3}`)

### ‚úÖ Escalares

Los **escalares** son tipos de datos que representan un solo valor indivisible, como n√∫meros, texto o booleanos.

En Python, los **tipos escalares** representan un solo valor, no estructuras. Son:

- **int**: Enteros (`10`)
- **float**: Decimales (`3.14`)
- **bool**: Booleanos (`True`, `False`)
- **str**: Texto (`"Hola"`)

Son at√≥micos: no se pueden dividir en partes m√°s peque√±as dentro del lenguaje.

```python

a = 10         # int
b = 3.14       # float
c = "Hola1"     # str
d = True       # bool
e = None       # NoneType

```

üß† `None` = ausencia expl√≠cita de valor (como `null` en otros lenguajes).

---

### ‚úÖ Contenedores

**Contenedores** en Python son tipos de datos que almacenan m√∫ltiples elementos, como listas, tuplas, diccionarios y conjuntos.

- **lista**: mutable, ordenada, permite cambios.
- **tupla**: inmutable, ordenada, no se puede modificar.
- **diccionario**: clave-valor, mutable, no ordenado (antes de Python 3.7).
- **conjunto**: elementos √∫nicos, no ordenado, mutable.

```python

lista = [1, 2, 3]            # mutable
tupla = (1, 2, 3)            # inmutable
diccionario = {"a": 1}       # clave-valor
conjunto = {1, 2, 3}         # sin duplicados

```

‚ö†Ô∏è `set` no garantiza orden.

## ‚öôÔ∏è **3. Operadores**

### ‚úÖ Aritm√©ticos y comparaci√≥n

En Python, los **operadores** son s√≠mbolos que realizan operaciones sobre valores o variables. Principales tipos:

- **Aritm√©ticos**: `+`, , , `/`, `%`, `*`, `//`
- **Relacionales**: `==`, `!=`, `>`, `<`, `>=`, `<=`
- **L√≥gicos**: `and`, `or`, `not`
- **Asignaci√≥n**: `=`, `+=`, `=`, `=`, `/=`, etc.
- **Bit a bit**: `&`, `|`, `^`, `~`, `<<`, `>>`
- **Membres√≠a**: `in`, `not in`
- **Identidad**: `is`, `is not`

```python

x = 5 + 3       # 8
y = 10 / 2      # 5.0
z = 10 // 3     # 3 ‚Üí divisi√≥n entera
mod = 10 % 3    # 1
exp = 2 ** 3    # 8

```

### ‚úÖ Comparaci√≥n

Los operadores de **comparaci√≥n** en Python se usan para comparar valores y devuelven `True` o `False`:

- `==` (igual a)
- `!=` (diferente de)
- `>` (mayor que)
- `<` (menor que)
- `>=` (mayor o igual que)
- `<=` (menor o igual que)

```python

a = 5
print(a == 5)   # True
print(a is 5)   # True en CPython, pero cuidado: `is` compara identidades, no valores

* `==` compara **valor**.
* `is` compara **identidad en memoria** (mismo objeto).

en este ejemplo, ambos dan `True` porque Python reutiliza objetos peque√±os (`int` entre -5 y 256), pero no siempre es seguro usar `is` para comparar valores.

```

üö´ No uses `is` para comparar valores num√©ricos o strings.

## üîÅ **4. Control de flujo**

El **control de flujo** en Python dirige la ejecuci√≥n del programa seg√∫n condiciones o repeticiones. Incluye:

- **if / elif / else**: decisiones condicionales
- **for**: bucle sobre secuencias
- **while**: bucle basado en condici√≥n
- **break** y **continue**: control dentro de bucles
- **try / except**: manejo de errores

### ‚úÖ Condicional

Una **condicional** es una estructura que ejecuta c√≥digo seg√∫n si una condici√≥n es verdadera o falsa.

```python
x = 10
if x > 5:
    print("Mayor")
elif x == 5:
    print("Igual")
else:
    print("Menor")
este c√≥digo es un ejemplo b√°sico de condicional en Python. Eval√∫a `x` y muestra:

* **"Mayor"** si `x > 5`
* **"Igual"** si `x == 5`
* **"Menor"** en cualquier otro caso

```

### ‚úÖ Bucle for


## üß∞ **5. Funciones**

Las **funciones** en Python son bloques de c√≥digo reutilizables que realizan una tarea espec√≠fica y pueden recibir y devolver valores.

### ‚úÖ B√°sica

Una funci√≥n b√°sica es un bloque de c√≥digo que realiza una tarea espec√≠fica y puede recibir datos de entrada y devolver un resultado.

- **def**: palabra clave para definir una funci√≥n.
- **funci√≥n**: bloque de c√≥digo reutilizable que puede recibir par√°metros y devolver resultados.
- **par√°metro**: variable dentro de la funci√≥n que recibe un valor al llamarla (ej. `nombre`, `num1`, `num2`).
- **return**: indica el valor que la funci√≥n devuelve cuando termina.
- **f-string**: formato de cadena para insertar variables en texto (ej. `f"Hola, {nombre}"`).
- **print()**: funci√≥n que muestra informaci√≥n en pantalla.

```python
def saludar(nombre):
    return f"Hola, {nombre}"

print(saludar("Uno"))

def sumar(num1, num2):
    return num1 + num2

print(sumar(5, 5))  # Salida: 10

    
def restar(nombre):
    return f"Hola, {nombre}"
    

```

### ‚úÖ Con valores por defecto

Una funci√≥n con valores por defecto tiene par√°metros que usan un valor inicial si no se proporciona uno al llamarla.

```python
def sumar(a, b=1):
    return a + b

esta funci√≥n suma `a` y `b`, donde `b` vale 1 si no se pasa al llamar:

sumar(5)    # devuelve 6 (5 + 1)  
sumar(5, 3) # devuelve 8 (5 + 3)

```

### ‚úÖ args y kwargs

- **`args`**: recibe m√∫ltiples argumentos posicionales como una tupla.
- **`kwargs`**: recibe m√∫ltiples argumentos nombrados como un diccionario.

```python
def sumar_todo(*args):
    return sum(args)

def imprimir_info(**kwargs):
    print(kwargs)

sumar_todo(1,2,3)             # args = (1,2,3)
imprimir_info(a=1, b=2)       # kwargs = {'a':1, 'b':2}

* `sumar_todo(*args)` recibe cualquier cantidad de valores y los suma.
* `imprimir_info(**kwargs)` recibe pares clave-valor y los muestra como diccionario.

```

üìå `*args` ‚Üí tupla de argumentos; `**kwargs` ‚Üí diccionario de argumentos nombrados.
